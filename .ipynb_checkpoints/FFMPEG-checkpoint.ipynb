```python
import os
import subprocess
import time
from os import listdir
import os
from os.path import isfile, join
import shutil

dir_name = "/Users/sjegan/Downloads/concat"
dest_dir_name = "/Users/sjegan/Downloads/dest"
bin_dir_name = "/Users/sjegan/Downloads/dest/del"

base_dir = dir_name
video_files = "/Users/sjegan/Downloads/video_list.txt"
output_file = "/Users/sjegan/Downloads/output.mp4"


files = []
onlyfiles = [(join(dir_name, f), f) for f in listdir(dir_name) if isfile(join(dir_name, f))]
onlyfiles.sort()
for loc, filename in onlyfiles:
    if loc.endswith(".mp4"):
        files.append(loc)
files.sort()

file_list = open(video_files, "w")
for f in files:
    file_list.write("file '%s'\n" % f)
file_list.close()

# merge the video files
cmd = ["ffmpeg",
       "-f",
       "concat",
       "-safe",
       "0",
       "-loglevel",
       "quiet",
       "-i",
       "%s" % video_files,
       "-c",
       "copy",
       "%s" % output_file
       ]




p = subprocess.Popen(cmd, stdin=subprocess.PIPE)

fout = p.stdin
fout.close()
p.wait()

print(p.returncode)


```

```python
def resizeImage(img):
    from PIL import Image
    image = Image.open(img)
    h = 600
    w = int((h * image.width) / image.height)
    image.thumbnail((w, h), Image.BILINEAR)
    image.save(img, 'JPEG', quality = 60)

import glob
convert = "/Users/sjegan/Downloads/com2/"


os.chdir(convert)
result = []
for file in glob.glob("**/*.jpg", recursive=True):
    result.append((convert + file, file[0:-4]))
for file in glob.glob("**/*.jpeg", recursive=True):
    result.append((convert + file, file[0:-4]))
for file in glob.glob("**/*.gif", recursive=True):
    result.append((convert + file, file[0:-4]))
for file in glob.glob("**/*.bmp", recursive=True):
    result.append((convert + file, file[0:-4]))
for file in glob.glob("**/*.png", recursive=True):
    result.append((convert + file, file[0:-4]))

i=1
try:
    for (filename, title) in result:
        filename_dest = filename
        import subprocess

#         os.chdir("C:\\Program Files\\IrfanView\\")
#         cmd= f'"C:\\Program Files\\IrfanView\\i_view64.exe" "{filename}" /advancedbatch /convert="{filename_dest}"'
#         subprocess.check_call(f'"i_view64" {filename}" /advancedbatch /convert="{filename_dest}"')
        # print(filename_dest)
        i=i+1
#         print(filename)
        resizeImage(filename)
        # break
finally:
    print(f'files={len(result)} converted={i}')
```

```python
from bs4 import BeautifulSoup
import urllib3
import re


def getLinks(url):
    http = urllib3.PoolManager()

    url = 'http://dl.new1music.ir/Serial/Foreign/Breaking%20Bad/Season1/'
    response = http.request('GET', url)
    print(response.text)
    soup = BeautifulSoup(response.data)
    links = []

    for link in soup.findAll('a', attrs={'href': re.compile("^http://")}):
        links.append(link.get('href'))

    return links

print( getLinks("https://arstechnica.com") )
```
